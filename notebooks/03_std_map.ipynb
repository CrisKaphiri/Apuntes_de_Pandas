## 2. Maps y Apply

#### 2.1 Maps

```python
# Uso de map para cambiar el valor de cada fila
df['country'].map(lambda row: 'Estados unidos' if row == 'United States' else row)
```

#### 2.2 Apply

```python
df['type'].unique()
```

```python
# Funcion para renombrar
def renombrar(row):
    if row['type'] == 'Movie':
        row['type'] = 'Pelicula'
    elif row['type'] == 'TV Show':
        row['type'] = 'Series de Televisión'
    return row
```

```python
df.apply(renombrar, axis = 1)
```
