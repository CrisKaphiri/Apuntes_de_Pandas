```python
# Dataframe de planetas

# Definir los datos de los planetas (nombre, radio en kilómetros, número de lunas)
data = {
    "Planeta": ["Mercurio", "Venus", "Tierra", "Marte", "Júpiter", "Saturno", "Urano", "Neptuno"],
    "Radio_km": [2439.7, 6051.8, 6371, 3389.5, 69911, 58232, 25362, 24622],
    "Lunas": [0, 0, 1, 2, 80, 83, 27, 14],
    "Tipo": ['Terrestre', 'Terrestre', 'Terrestre', 'Terrestre', 'Gigante de gas', 'Gigante de gas', 'Gigante de Hielo', 'Gigante de Hielo'],
    "Anillos": ['no', 'no', 'no', 'no', 'si', 'si', 'si', 'si'],
    "Temperatura_media_c": [167, 464, 15, -65, -110, -140, -195, -200],
    "Campo_magnetico": ['si', 'no', 'si', 'no', 'si', 'si', 'si', 'si']
}

# Crear el DataFrame
planetas = pd.DataFrame(data)

# Mostrar el DataFrame
print(planetas)
```

```python
# Planetas que tengan menos de 20 Lunas
planetas[planetas["Lunas"] < 20]
```

```python
# Planetas que tengan menos de 10 Lunas o más de 50 Lunas
planetas[(planetas["Lunas"] < 10) | (planetas["Lunas"] > 50)]
```

```python
# Planetas con mas de 20 lunas, pero no planetas que tengan 80 ni radio inferior a 50000km
planetas[(planetas["Lunas"] > 20) & ~(planetas["Lunas"] == 80) & ~(planetas["Radio_km"] < 50000)]
```

```python
# Solamente filtrar
planetas.loc[(planetas["Lunas"] > 20) & ~(planetas["Lunas"] == 80) & ~(planetas["Radio_km"] < 50000), "Planeta"]
```

```python
# Groupby
planetas.groupby(['Tipo']).sum(numeric_only=True)
```

```python
# Groupby
planetas.groupby(['Tipo', 'Campo_magnetico']).mean(numeric_only= True)
```

```python
# Agrupar planetas por su tipo y calcular valores medios y medianos
planetas.groupby(['Tipo'])[planetas.select_dtypes(include=['number']).columns].agg(['mean', 'median'])
```
