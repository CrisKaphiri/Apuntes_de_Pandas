## 3 - Selección mediante etiquetas con loc[filas, columnas]

#### 3.1 - Obtener el primer valor de la columna especifica

```python
df.loc['s1', 'country']
```

#### 3.2 - Obtener varias columnas de un dataframe

```python
df.loc[:, ['title', 'duration', 'description']]
```

#### 3.3 - Acceder a filas mediante condiciones

```python
df.loc[df['country'] == 'Chile']
```

```python
df.loc[(df['country'] == 'Chile') & (df['type'] == 'Movie')]
```

#### 3.4 - Acceder a varios valores de una columna

```python
df.loc[df['country'].isin(['Chile', 'United States'])].sort_values(by = 'country')
```

#### 3.5 - Acceder a valores no nulos

```python
df.loc[df['director'].notnull()]
```

#### 3.6 - Acceder a valores nulos

```python
df.loc[df['director'].isnull()]
```

#### 3.7 - Todos los valores nulos
