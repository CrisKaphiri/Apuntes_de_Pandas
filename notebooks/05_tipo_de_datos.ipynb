```python
import pandas as pd
```

```python
df = pd.read_csv("../data/netflix_titles.csv")
```

## 1 - Tipos de datos

#### 1.1 - Ver tipo de dato en cada columna

```python
df.dtypes
```

#### 1.2 - Cambiar el tipo de dato de int a str

```python
df['release_year'] = df['release_year'].astype(str)
```

#### 1.3 - Convierte a número forzando errores a NaN

```python
pd.to_numeric(df['release_year'], errors='coerce')
```

## 2 - Valores núlos

#### 2.1 - Suma valores nulos

```python
df.isnull().sum()
```

#### 2.2 - Devuelve un df con valores nulos

```python
df[pd.isnull(df['director'])]
```

#### 2.2 - Devuelve una serie booleana de nulos

```python
df['director'].isnull()
```

#### 2.3 - Elimina filas con valores nulos

```python
df.dropna()
```

####

#### 2.4 - Elimina filas si la columna tiene el valor nulo

```python
df.dropna(subset = ['director'])
```

## 3 - Rellenar valores nulos

#### 3.1 - Rellenar valores con fillna

```python
# Rellena valores NaN con 'Desconocido'
df['director'].fillna('Desconocido')
```

#### 3.2 - Rellena los valores NaN con el valor de la siguiente fila

```python
df['date_added'].bfill()
```

#### 3.3 - Rellena los valores NaN con el valor de la fila anterior

```python
df['date_added'].ffill()
```

#### 3.4 - Reemplaza un valor

```python
df['type'].replace('Movie', 'Peliculas')
```
